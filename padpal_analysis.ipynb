# Location Analysis

```python
import matplotlib.pyplot as plt

locations = df['common_location'].str.split(';').explode()

location_counts = locations.value_counts()

location_counts.plot(
    kind='barh',
    figsize=(8,5)
)

plt.title('Where Emergency Most Commonly Occurs')
plt.xlabel('Mentions')
plt.ylabel('Location')
plt.tight_layout()

plt.show()
```

### Insight
College/School emerged as the most frequently reported location for menstrual emergencies, followed by traveling and office environments.

# Key Insights

- 97% of respondents have faced a menstrual emergency.
- 92% would use the proposed platform.
- 93% are willing to help another woman.
- College campuses are the most common emergency location.
- Users prefer an all-in-one solution rather than a single feature.
